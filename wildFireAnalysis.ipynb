```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('California Wildfire Damage.csv')
data.head()
```

Cleaning the Data

```python
data = data.drop_duplicates()

#Rename columns and add month/date columns
data['Date'] = pd.to_datetime(data['Date'])
data.rename(columns={"Area_Burned (Acres)": "AreaBurned"}, inplace=True)
data.rename(columns={"Homes_Destroyed": "HomesDestroyed"}, inplace=True)
data.rename(columns={"Businesses_Destroyed": "BusinessDestoryed"}, inplace=True)
data.rename(columns={"Vehicles_Damaged": "VehiclesDamaged"}, inplace=True)
data.rename(columns={"Estimated_Financial_Loss (Million $)": "FinancialLoss"}, inplace=True)
data['Month'] = data['Date'].dt.month
data['Year'] = data['Date'].dt.year


#Fill missing values for numerical values with the mean.
numerical_cols = data.select_dtypes(include=['number']).columns
data[numerical_cols] = data[numerical_cols].fillna(data[numerical_cols].mean())

#Fill missing values for categorical values with the mode.
categorical_cols = data.select_dtypes(include=['category','object']).columns
data[categorical_cols] = data[categorical_cols].fillna(data[categorical_cols].mode().iloc[0])
```
